```python
import json
import pytrec_eval
```

```python
f = open('res10000.json', 'r')
g = open('qrels.json', 'r')
```

```python
run = json.load(f)
qrels = json.load(g)
```

```python
evaluator = pytrec_eval.RelevanceEvaluator(qrels, {'P_1'})
x = evaluator.evaluate(run)
print (x)
```

```python
p_1= []
for i in x.items():
    p_1.append((i[1]['P_1']))
```

```python
sum(p_1)/len(p_1)
```
